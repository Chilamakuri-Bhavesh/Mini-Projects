```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
```

```python
asia= pd.read_excel("ASIA.xlsx")
```

```python
asia.dropna(inplace=True)
```

```python
europe=pd.read_excel("EUROPE.xlsx")
```

```python
asia
```

```python
europe
```

```python
label_encoder = LabelEncoder()
```

```python
asia["Country"]=label_encoder.fit_transform(asia['Country'])
```

```python
asia
```

```python
europe["Country"]=label_encoder.fit_transform(europe['Country'])
```

```python
europe
```

```python
x_train= asia.iloc[:,:-1]
```

```python
x_train
```

```python
y_train= asia.iloc[:,-1:]
```

```python
y_train
```

```python
x_test= europe.iloc[:,:-1]
```

```python
x_test
```

```python
y_test= europe.iloc[:,-1:]
```

```python
y_test
```

```python
dt_regressor = DecisionTreeRegressor()
```

```python
dt_regressor.fit(x_train, y_train)
```

```python
y_pred = dt_regressor.predict(x_test)
```

```python
y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("mse = ",mse)
print("r2= ", r2)
```
